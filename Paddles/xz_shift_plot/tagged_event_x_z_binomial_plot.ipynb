```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_tagged_event = []
x_axis = []

z_tagged_event = []
z_axis = []

with open('x_100_109.txt') as file_x:
    for line in file_x:
        data = line.split()
        x_axis.append(int(data[0]))
        x_tagged_event.append(int(data[1]))

x_fraction = [a / 500. for a in x_tagged_event]
x_axis
x_fraction
with open('z_100_109.txt') as file_z:
    for line in file_z:
        data = line.split()
        z_axis.append(int(data[0]))
        z_tagged_event.append(int(data[1]))
z_fraction = [a / 500. for a in z_tagged_event]
```

```python
x_axis
plt.grid(True)
p1 = plt.plot(x_axis, x_fraction,  color = 'b',marker = '*')
p2 = plt.plot(z_axis, z_fraction,  color = 'r',marker = 'o')
plt.legend( (p1[0], p2[0]), ('X_shift', 'Z_shift') )

#plt.figure()
```
