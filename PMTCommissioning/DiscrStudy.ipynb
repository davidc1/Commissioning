```python
%matplotlib inline
#%pylab
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mpc
import matplotlib.dates as dts
import numpy as np
import pandas as pd
import itertools
import os
import ROOT
import datetime
from root_numpy import root2array, root2rec, tree2rec, array2root
plt.rcParams.update({'font.size': 16})
```

```python
f = ROOT.TFile('opdetwf_tree.root')
t = f.Get("_tree")
t.Show(0)
arr = tree2rec(t,branches=['_ev','_ch','_adcs','_adc_v','_tstart','_tend'])
df = pd.DataFrame(arr)
```

```python
ev_arr = []
ch_arr = []

from matplotlib.colors import LogNorm

for x in xrange(len(arr)):
    evt = arr['_ev'][x]
    chn = arr['_ch'][x]
    samples = arr['_adcs'][x]
    if ( (evt < 102) and (chn < 34) ):
        if (samples == 1001):
            ev_arr.append(evt)
            ch_arr.append(chn)

            

fig = plt.figure(figsize=(15,7))
plt.hist2d(ev_arr,ch_arr,bins=(101,33))#,norm=LogNorm())
plt.colorbar()
plt.xlim([0,100])
plt.ylim([0,31])
plt.xlabel('Event Number')
plt.ylabel('Channel Number')
plt.title('Number of 1000-tick Readout Windows')
plt.show()
    
```

```python
n = 0
m = 0
while (n < 20):
    if (arr['_adcs'][m] == 21):
        fig = plt.figure(figsize=(10,6))
        plt.plot(arr['_adc_v'][n],'b-',lw=2)
        plt.grid()
        evt = arr['_ev'][m]
        chn = arr['_ch'][m]
        plt.title('Waveform for Chan %i, Event %i'%(chn,evt))
        plt.xlabel('Tick [ 64 MHz sampling]')
        plt.ylabel('ADC Count')
        plt.show()
        n+= 1
    m += 1
```
